```python
def file_prop(file_path):
    with open(file_path,'r') as f:
        lines=f.readlines()
        line_count=len(lines)
        word_count=sum([len(line.split()) for line in lines])
        char_count=sum(len(line) for line in lines)
        char_count_non_space=sum([len(word) for line in lines for word in line.split()])
        return line_count,word_count,char_count,char_count_non_space

a,b,c,d=file_prop("example.txt")
print(a,b,c,d)
```

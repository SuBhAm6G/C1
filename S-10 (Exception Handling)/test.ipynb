```python
try:
    a=int(input("Enter a num: "))
    res=34/a
except ZeroDivisionError as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e:
    print(e)

else:
    print(f"The result is {res}")

```

```python
try:
    a=int(input("Enter a num: "))
    res=34/a
except ZeroDivisionError as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e:
    print(e)

else:
    print(f"The result is {res}")
finally:
    print("Execution Completed")
```
